```python
import numpy as np
import cv2 as cv
from PIL import Image
from scipy import ndimage
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = np.array(Image.open("Berkeley.jpg"))
Image.fromarray(img)
```

```python

# Using Gamma correction to adjust image dark and bright contrast level
def adjust_gamma(image, gamma=1.0):
    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255
        for i in np.arange(0, 256)]).astype("uint8")
    return cv.LUT(image, table)
```

```python
adjusted = adjust_gamma(img, gamma=0.7)
test_img1 = Image.fromarray(adjusted)
test_img1.save("Gamma_Corrected_Berkeley.png")
display(Image.fromarray(adjusted))
```

```python
adjusted = adjust_gamma(img, gamma=0.6)
test_img1 = Image.fromarray(adjusted)
test_img1.save("Gamma_Corrected_Berkeley.png")
display(Image.fromarray(adjusted))
```

I implemented Gamma correction on images with gamma value equals to 0.7 and 0.6. I chose  this value because this value in particular was showing details with equal contrast and decreased brightness value. 
0.7 had a better looking image output.
Steps
Import image using PIL Image module and converted to numpy array
Now made a function for gamma correction which takes input image and divide each value to convert it from [0,1] and then multiplied with desired value called gamma corrected.
Now given the input image it generates an image that can be viewed and saved using PIL package.
