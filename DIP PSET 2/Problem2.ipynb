```python
from IPython.display import display, Math, Latex
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = Image.open("spillway-dark.tif")
img = np.asarray(img)
plt.imshow(img, cmap='gray')
```

### **Part a**

```python
def get_histogram(image, bins):
    histogram = np.zeros(bins)
    for pixel in image:
        histogram[pixel] += 1
    return histogram
```

```python
def cumsum(a):
    a = iter(a)
    b = [next(a)]
    for i in a:
        b.append(b[-1] + i)
    return np.array(b)
```

```python
def Hist_Eq(img, bins):
    image = img.flatten()
    hist = get_histogram(image, bins)
    plt.plot(hist)
    plt.show()
    cs = cumsum(hist)
    plt.plot(cs)
    plt.show()
    nj = (cs - cs.min()) * 255
    N = cs.max() - cs.min()
    cs = nj / N
    cs = cs.astype('uint8')
    plt.plot(cs)
    plt.show()
    
    img_new = cs[image]
    plt.hist(img_new, bins=50)
    
    img_new = np.reshape(img_new, img.shape)
    fig = plt.figure()
    fig.set_figheight(15)
    fig.set_figwidth(15)
    fig.add_subplot(1,2,1)
    plt.imshow(img, cmap='gray')
    fig.add_subplot(1,2,2)
    plt.imshow(img_new, cmap='gray')
    plt.show(block=True)
    
    
```

### **Part b**

```python
Hist_Eq(img, 256)
```

### **Part c**

```python
img1 = Image.open("hidden-horse.tif")
img1 = np.asarray(img1)
plt.imshow(img1, cmap='gray')
```

```python
Hist_Eq(img1, 256)
```
